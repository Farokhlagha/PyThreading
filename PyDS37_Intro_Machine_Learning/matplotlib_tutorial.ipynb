```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.__version__
```

```python
#Draw a line in a diagram from position (0,0) to position (6,250):
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Draw a line in a diagram from position (1,3) to position (8,10):
x2points = np.array([1, 8])
y2points = np.array([3, 10])

plt.plot(x2points, y2points)
plt.show()
```

```python
#without line
x3points = np.array([1, 8])
y3points = np.array([3, 10])

plt.plot(x3points, y3points, 'o')
plt.show()
```

```python
#special points
x4points = np.array([1, 2, 6, 8])
y4points = np.array([3, 8, 1, 10])

plt.plot(x4points, y4points)
plt.show()
```

```python
#plot with default x point
y5points = np.array([3, 8, 1, 10, 5, 7])

plt.plot(y5points)
plt.show()
```

```python
#Marker
y6points = np.array([3, 8, 1, 10])

plt.plot(y6points, marker = 'o')
plt.show()
```

```python
#Star marker
y7points = np.array([3, 8, 1, 10])

plt.plot(y7points, marker = '*')
plt.show()
```

```python
#Diamond marker
y8points = np.array([3, 8, 1, 10])

plt.plot(y8points, marker = 'D')
plt.show()
```

```python
#fmt
y9points = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'H-.y')
plt.show()
```

```python
y10points = np.array([3, 8, 1, 10])

plt.plot(y10points, marker = '*', ms = 20, mec="m", mfc="y")
plt.show()
```

```python
plt.plot(y10points, marker = 'o', ms = 20, mec='#4CAF50', mfc='hotpink')
plt.show()
```

```python
plt.plot(y10points, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(y10points, linestyle = 'dashed')
```
